```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = './data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
app_train.head()
```

```python
app_train.describe()
```

```python
# 抓取'DAYS_EMPLOYED'欄位
avg = app_train['DAYS_EMPLOYED'].mean()
print("\'DAYS_EMPLOYED\'欄位的平均數: {}".format(avg))
std = app_train['DAYS_EMPLOYED'].std()
print("\'DAYS_EMPLOYED\'欄位的標準差: {}".format(std))
```

```python
import matplotlib.pyplot as plt
n_split = 100 # 分割成幾個區間
n,bins,patches = plt.hist(app_train['DAYS_EMPLOYED'], n_split)
```

```python
# 抓取'AMT_GOODS_PRICE'欄位
avg = app_train['AMT_GOODS_PRICE'].mean()
print("\'AMT_GOODS_PRICE\'欄位的平均數: {}".format(avg))
std = app_train['DAYS_EMPLOYED'].std()
print("\'AMT_GOODS_PRICE\'欄位的標準差: {}".format(std))
```

```python
n_split = 100 # 分割成幾個區間
n,bins,patches = plt.hist(app_train['AMT_GOODS_PRICE'], n_split)
```
